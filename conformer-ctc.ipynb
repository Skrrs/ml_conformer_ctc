```python
import nemo
import nemo.collections.asr as nemo_asr # Nemo ASR class
from nemo.utils import exp_manager # 학습 로그, conf등에 사용되는 lib
import nlptutti as metrics
```

```python
Con_model = nemo_asr.models.ASRModel.restore_from("Conformer-CTC-BPE-8000-KOR.nemo")
```

```python
def stt(audio_file, ref, model):
    test_str_list = []
    test_str_list.append(audio_file)
    result = Con_model.transcribe(test_str_list)
    pred = result[0]
    _cer = metrics.get_cer(ref, pred)['cer']
    
    return pred, _cer*100
```

```python
# 정답 text
ref1 = "한국에 온 지 벌써 이년 됐어요. 미리 잘 모르지만 아마 내년에까지 여기 있을 거예요."
# audio 파일 경로
audio_file1 = "audio.wav"

# 정답 text
ref2 = "안녕하세요 제 이름은 허승회입니다."
# audio 파일 경로
audio_file2 = "test.mp3"

# model_(version) --> version에 b,s,m,l 삽입
stt(audio_file2, ref2, Con_model)
```
